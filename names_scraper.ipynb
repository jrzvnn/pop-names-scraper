```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://forebears.io/philippines/forenames'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

### Item of interest:
* Rank
* Gender
* Gender Percentage
* Forename
* Incidence
* Frequency

```python
forename = []
incidence = []
frequency = []
gender = []
gender_percentage = []

table = soup.find('tbody')
rows = table.find_all('tr')

# Scraping forename
for row in rows:
    td_elements = row.find_all('td')
    if len(td_elements) == 5:
        # Extract the forename from the third <td> element
        forename_element = row.find_all('td')[2].find('a')
        forename_value = forename_element.get_text(strip=True)
        forename.append(forename_value)

# Scraping incidence
for row in rows:
    td_elements = row.find_all('td')
    if len(td_elements) == 5:
        # Extract the incidence from the fourth <td> element
        incidence_element = row.find_all('td')[3]
        incidence_value = incidence_element.get_text(strip=True)
        incidence.append(incidence_value)

# Scraping frequency
for row in rows:
    td_elements = row.find_all('td')
    if len(td_elements) == 5:
        # Extract the frequency from the fifth <td> element
        frequency_element = row.find_all('td')[4].find('span')
        frequency_value = frequency_element.get_text(strip=True)
        frequency.append(frequency_value)

# Scraping gender
for row in rows:
    td_elements = row.find_all('td')
    if len(td_elements) == 5:
        # Check the gender based on the presence of <div class="m"> element
        element = row.find_all('td')[1].find('div', class_='m')
        if element and '%' in element.get_text(strip=True):
            gender.append('Male')
        else:
            gender.append('Female')

# Scraping gender percentage
for row in rows:
    td_elements = row.find_all('td')
    if len(td_elements) == 5:
        # Extract the gender percentage based on the conditions
        m_div = td_elements[1].find('div', class_='m')
        f_div = td_elements[1].find('div', class_='f')
        if m_div and '%' in m_div.get_text(strip=True):
            gender_percentage.append(m_div.get_text(strip=True).split('%')[0])
        elif f_div and '%' in f_div.get_text(strip=True):
            gender_percentage.append(f_div.get_text(strip=True).split('%')[0])
```

```python
import csv

# Define the file path for the CSV file
csv_file_path = 'scraped_data.csv'

# Create a CSV file and write the headers
with open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['rank', 'forename', 'incidence', 'frequency', 'gender', 'gender_percentage'])

    # Write the data rows
    for i in range(len(forename)):
        rank = i + 1  # Calculate the rank as the index + 1
        writer.writerow([rank, forename[i], incidence[i], frequency[i], gender[i], gender_percentage[i]])
```
